```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
im = cv2.imread("skeleton.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(im, 'gray')
```

```python
Laplacian = cv2.Laplacian(im,cv2.CV_16S,ksize=3)
plt.imshow(Laplacian, 'gray')
cv2.imwrite('1Laplacian.png',Laplacian)
```

```python
newIm = im + Laplacian
plt.imshow(newIm, 'gray')
cv2.imwrite('2newIm.png',newIm)
```

```python
sobelx = cv2.Sobel(im, cv2.CV_16S, 1, 0, ksize=3)
sobely = cv2.Sobel(im, cv2.CV_16S, 0, 1, ksize=3)
sobel = cv2.convertScaleAbs(sobelx)+cv2.convertScaleAbs(sobely) 
plt.imshow(sobel, 'gray')
cv2.imwrite('3sobel.png',sobel)
```

```python
blurredIm = cv2.blur(sobel, (5, 5))
plt.imshow(blurredIm, 'gray')
cv2.imwrite('4blurredIm.png',blurredIm)
```

```python
maskIm = cv2.multiply(newIm,blurredIm/255,dtype = cv2.CV_16S)
plt.imshow(maskIm, 'gray')
cv2.imwrite('5maskIm.jpg', maskIm)
```

```python
newIm2 = im + maskIm
plt.imshow(newIm2, 'gray')
cv2.imwrite('6newIm2.png',newIm2)
```

```python
img = np.array(255*(newIm2/255)**0.5)
plt.imshow(img, 'gray')
cv2.imwrite('7img.png',img)
```
